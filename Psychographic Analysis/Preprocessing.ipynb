```python
df= pd.read_csv('D:\\Class work\\2nd Year\\4th Sem\\Capstone\\Segmentation\\Actual_analysis\\LIWC\\Test2\\LIWC_for_OCEAN_Reviews.csv')
```

```python
#list(df.columns)
Openness=['percept','hear','excl']
Conscientiousness=['achieve','work','body','death','anger','excl']
Extraversion=['humans','social','family']
Agreeableness= ['certain', 'incl','family','body','anger']
Neuroticism=['sad','negemo','body','anger','home','anx','work']
traits=(Openness,Conscientiousness,Extraversion,Agreeableness,Neuroticism)
traits
```

```python
for trait in traits:
    df['Mean_'+str(trait)]=df[trait].mean(axis=1)
```

```python
df = df.rename(columns = {'Source (A)':'Text'})
df=df.drop(columns=df.columns[1:-5],axis=1)
col_name=['Openness','Conscientiousness','Extraversion','Agreeableness','Neuroticism']
for j in range(1,6):
    old = df.columns[j]
    new = col_name[j-1]
    df = df.rename(columns = {old:new})
```

```python
df=df.drop_duplicates()
OCEAN=df.sort_values('Text').reset_index(drop=True)
OCEAN.head()
```

# Polarity

```python
df= pd.read_csv('D:\\Class work\\2nd Year\\4th Sem\\Capstone\\Segmentation\\Actual_analysis\\MEM\\Reviews\\New folder\\2021-04-04_-_PCA_Results - Scores.csv')
```

```python
cols = df.columns.tolist()
cols = cols[-1:] + cols[:-1]
df = df[cols]
df = df.rename(columns = {'Filename':'Text'})
df=df.drop_duplicates()
Themes=df.sort_values('Text').reset_index(drop=True)
df=df.drop(columns=df.columns[1:],axis=1)
```

```python
from textblob import TextBlob
df['Text']= df['Text'].astype(str) #Make sure about the correct data type
pol = lambda x: TextBlob(x).sentiment.polarity
df['Polarity'] = df['Text'].apply(pol)
df['Sentiment'] = df['Polarity'].apply(lambda x:'Good' if x>0 else ('Neutral' if x==0 else 'Bad')  )
```
